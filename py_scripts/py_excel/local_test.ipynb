```python
import openpyxl as xl
import pandas as pd
from copy import copy
import pinyin
```

```python
path = 'jhu_2022_12_30.xlsx'
wb = xl.load_workbook(filename = path)
wb_new = xl.Workbook()
```

```python
ws = wb['私募股票量化数据【管理人版】']
datasource = ws['A6':'P59']
row_len = len(datasource)
col_len = len(datasource[0])
```

```python
# for row,i  in datasource, range(row_len):
#     for cell, j  in row, range(col_len):
#         ws_new[i, j] = cell
ws_new = wb_new.create_sheet('new_sheet_2')
# ws_new = wb_new.create_sheet('new_sheet_4')
for i in range(row_len):
    for j in range(col_len):
        cell = datasource[i][j]
        ws_new.cell(i + 1, j + 1).value = cell.value
        if cell.has_style:
            ws_new.cell(i + 1, j + 1).font = copy(cell.font)
            ws_new.cell(i + 1, j + 1).border = copy(cell.border)
            ws_new.cell(i + 1, j + 1).fill = copy(cell.fill)
            ws_new.cell(i + 1, j + 1).number_format = copy(cell.number_format)
            ws_new.cell(i + 1, j + 1).protection = copy(cell.protection)
            ws_new.cell(i + 1, j + 1).alignment = copy(cell.alignment)
```

```python
wb_new.save('temp.xlsx')
```

```python
data = ws_new.values
data = list(data)
cols = data[0]
data = data[1:]
df = pd.DataFrame(data, columns=cols)
df
```

```python
arr_initial_pinyin = []
for name in df.iloc[:, 0]:
    initial_pinyin = pinyin.get_initial(name, delimiter='').upper()
    arr_initial_pinyin.append(initial_pinyin)
    print(f"{name} {initial_pinyin}")
```

```python
df['拼音首写'] = arr_initial_pinyin

cols = list(cols)
cols.insert(0, '拼音首写')

df = df[cols]
```

```python
df.head(5)
```

```python
ws_max_draw_back = wb_new.create_sheet('最大回撤')
```

```python


from openpyxl.utils.dataframe import dataframe_to_rows

cols_max_draw_back = ['拼音首写', '阿尔法产品', '2018年以来最大回撤']
rows = dataframe_to_rows(df[cols_max_draw_back], index=False, header=True)
```

```python
for r in rows:
    ws_max_draw_back.append(r)
    # print(r)
    
for cell in ws['A'] + ws[1]:
    cell.style = 'Pandas'
```

```python
wb_new.save('temp.xlsx')
```

### 画图

```python
from openpyxl.chart import (
    AreaChart,
    Reference,
    Series,
)
```

```python
chart = AreaChart()
chart.title = "Area Chart"
chart.style = 13
chart.x_axis.title = 'Test'
chart.y_axis.title = 'Percentage'

cats = Reference(ws_max_draw_back, min_col=1, min_row=1, max_row=54)
data = Reference(ws_max_draw_back, min_col=2, min_row=1, max_col=3, max_row=54)
chart.add_data(data, titles_from_data=True)
chart.set_categories(cats)

ws_max_draw_back.add_chart(chart, "F10")

wb_new.active = ws_max_draw_back
wb_new.save("temp.xlsx")
```
